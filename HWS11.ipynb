```python
from matplotlib import pyplot 
import math

def f(x):
    return -12*(x**4)*math.sin(math.cos(x)) - 18*(x**3)+5*(x**2) + 10*x - 30

x = [x/10 for x in range(-30, 31)]
y = [f(x) for x in x]
pyplot.plot(x, y, 'r-')
```

```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')
p1 = -12*(x**4)*sin(cos(x)) - 18*(x**3)+5*(x**2) + 10*x - 30
plot(p1)
```

```python
solve(p1, x)
```
